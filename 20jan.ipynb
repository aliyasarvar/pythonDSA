```python
#3
temp = float(input("Enter body temperature: "))

if temp >= 85.0 and temp <= 91.0:
    print("Serious Hypothermia")

elif temp > 91.0 and temp < 95.0:
    print("Mild Hypothermia")

elif temp >= 95.0 and temp <= 98.0:
    print("Normal Temperature")

elif temp > 98.0 and temp <= 100.0:
    print("Mild Fever")

elif temp > 100.0 and temp <= 105.0:
    print("High Fever")

else:
    print("Temperature out of range")
```

```python
#4
n = int(input("Enter a number: "))

if n > 0 and n % 2 != 0:
    print("Number is Positive and Odd")

elif n > 0 and n % 2 == 0:
    print("Number is Positive and Even")

elif n < 0 and n % 2 != 0:
    print("Number is Negative and Odd")

elif n < 0 and n % 2 == 0:
    print("Number is Negative and Even")

else:
    print("Number is Zero")
```

```python
#9
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a Prime Number")
else:
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    
    if count == 2:
        print("Prime Number")
    else:
        print("Not a Prime Number")
```
