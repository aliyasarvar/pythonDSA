```python
#3
def sumn(n):
    if n == 1:
        return 1
    return n + sumn(n-1)

print(sumn(10))
```

```python
#4
def sum_array(arr):
    if len(arr) == 0:
        return 0
    return arr[0] + sum_array(arr[1:])

arr = [1,2,3,4]
print(sum_array(arr))
```

```python
#5
def search(arr, x):
    if len(arr) == 0:
        return False
    if arr[0] == x:
        return True
    return search(arr[1:], x)

arr = [1,2,3,4]
print(search(arr, 3))
```

```python
#6
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

```python
#7
def sorted(arr):
    if len(arr) <= 1:
        return True
    if arr[0] > arr[1]:
        return False
    return sorted(arr[1:])

arr = [1,2,3,4]
print(sorted(arr))
```

```python
#8
def is_sorted(arr, i=0):
    if i == len(arr)-1:
        return True
    if arr[i] > arr[i+1]:
        return False
    return is_sorted(arr, i+1)

arr = [1,2,3,4]
print(is_sorted(arr))
```

```python
#9
def min_max(arr):
    minimum = maximum = arr[0]

    for i in arr[1:]:
        if i < minimum:
            minimum = i
        elif i > maximum:
            maximum = i

    return minimum, maximum

arr = [3,5,1,8,2]
print(min_max(arr))
```

```python
#10
def reverse_array(arr):
    left = 0
    right = len(arr)-1

    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1

    return arr

arr = [1,2,3,4]
print(reverse_array(arr))
```

```python
#11
def smallest(arr, k):
    arr.sort()
    return arr[k-1]

arr = [7,10,4,3,20,15]
print(smallest(arr, 3))
```

```python
#12
def sort012(arr):
    low = mid = 0
    high = len(arr)-1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr

arr = [0,2,1,2,0,1]
print(sort012(arr))
```
