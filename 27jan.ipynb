```python
#1
def countchar(s, ch):
    count = 0
    for i in s:
        if i == ch:
            count += 1
    return count

string = input("Enter a string: ")
char = input("Enter a character: ")

print("Count is:", countchar(string, char))
```

```python
#2
def replacevowels(s):
    vowels = "aeiouAEIOU"
    result = ""
    
    for i in s:
        if i in vowels:
            result += "*"
        else:
            result += i
    return result

string = input("Enter a string: ")
print(replacevowels(string))
```

```python
#4
def reverse_string(s):
    rev = ""
    for i in range(len(s)-1, -1, -1):
        rev += s[i]
    return rev

string = input("Enter a string: ")
print(reverse_string(string))
```

```python
#5
def check_palindrome(s):
    rev = ""
    for i in range(len(s)-1, -1, -1):
        rev += s[i]
    
    if s == rev:
        return "Palindrome"
    else:
        return "Not Palindrome"

string = input("Enter a string: ")
print(check_palindrome(string))
```

```python
#6
def case(s):
    words = s.split()
    result = ""
    
    for w in words:
        result += w.capitalize() + " "
    
    return result.strip()

string = input("Enter a string: ")
print(case(string))
```

```python
#7
def deleteChar(s, ch):
    new_string = ""
    
    for i in s:
        if i != ch:
            new_string += i
    
    return new_string

string = input("Enter a string: ")
char = input("Enter character to delete: ")

print(deleteChar(string, char))
```

```python
#8
def sum_of_digits(s):
    total = 0
    
    for i in s:
        if i.isdigit():
            total += int(i)
    
    return total

string = input("Enter a string: ")
print("Sum of digits:", sum_of_digits(string))
```

```python
#1
def increment_list(lst):
    for i in range(len(lst)):
        lst[i] = lst[i] + 5
    return lst

numbers = [1, 2, 3, 4]
print(increment_list(numbers))
```

```python
#3
n = int(input("Enter number of elements: "))

lst = []
for i in range(n):
    num = int(input())
    lst.append(num)

positive = []
negative = []

for i in lst:
    if i >= 0:
        positive.append(i)
    else:
        negative.append(i)

print("Original list:", lst)
print("Positive list:", positive)
print("Negative list:", negative)
```

```python
#4
def find_largest(lst):
    largest = lst[0]
    for i in lst:
        if i > largest:
            largest = i
    return largest

numbers = [10, 45, 23, 89, 12]
print("Largest element:", find_largest(numbers))
```

```python
def second_largest(lst):
    lst.sort()
    return lst[-2]

numbers = [10, 45, 23, 89, 12]
print("Second largest:", second_largest(numbers))
```
