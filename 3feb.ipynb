```python
#1
def islice(arr):
    if len(arr) <= 1:
        return True
    if arr[0] > arr[1]:
        return False
    return isslice(arr[1:])
```

```python
#1
def unions(a, b):
    result = list(set(a + b))
    return result
a = [1, 2, 2, 3]
b = [2, 4, 5]
print(unions(a, b))
```

```python
#2
def one(arr):
    if len(arr) == 0:
        return arr
    return [arr[-1]] + arr[:-1]
arr = [1, 2, 3, 4, 5]
print(one(arr))
```
