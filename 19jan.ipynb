```python
#1
num = int(input("Enter a number: "))

temp = num
power = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** power
    temp = temp // 10

if sum == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```
